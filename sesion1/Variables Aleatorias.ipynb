```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
```

```python
N = 10000
x_normal = np.random.normal(scale = 2, loc = 1, size = N)
```

```python
plt.figure('Datos simulados de una Normal')
plt.hist(x_normal, bins = 50)
```

```python
norm.cdf(0, loc = 1, scale = 0.7)
```

```python
norm.ppf(0.05)
```

```python
x_exp = np.random.exponential(scale = 1/10, size = N)
```

```python
plt.figure('Histograma datos exponenciales')
plt.hist(x_exp)
```
